```python
import os
import google.generativeai as genai
import math
```

```python
GENAI_AI_KEY = '**'
```

```python
genai.configure(api_key=GENAI_AI_KEY)
```

```python
model_name = 'models/gemini-2.0-flash'
```

```python
model = genai.GenerativeModel(model_name, generation_config={'temperature':1.0})
response = model.generate_content('hello')
```

```python
print(response)
```

```python
avg_logprobs = response.candidates[0].avg_logprobs
confidence_score =(math.exp(avg_logprobs))
```

```python
print(confidence_score)
```
